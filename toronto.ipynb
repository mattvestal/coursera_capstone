```python
import pandas as pd
import numpy as np
import requests
import io
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
print('-----------------------------------------------------------')
from geopy.geocoders import Nominatim 
print('-----------------------------------------------------------')
!conda install -c conda-forge folium=0.5.0 --yes 
print('-----------------------------------------------------------')
import folium 

```

```python
URL = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
r = requests.get(URL)
```

```python
soup = BeautifulSoup(r.content, 'html5lib') 
#print(soup.prettify()) 
```

```python
mytable = soup.find('table',{'class':'wikitable'})
A=[]
B=[]
C=[]
for row in mytable.findAll('tr'):
    cells=row.findAll('td')
    if len(cells)==3:
        A.append(cells[0].find(text=True))
        B.append(cells[1].find(text=True))
        C.append(cells[2].find(text=True))

df=pd.DataFrame(A,columns=['postcode'])
df['borough']=B
df['neighborhood']=C
#df['Kit_manufacturer']=D
#df['Shirt_sponsor']=E
df = df[df.borough != 'Not assigned'] 
df = df[df.neighborhood != 'Not assigned']
df = df.reset_index(drop=True)
df = df.groupby(['postcode', 'borough'])['neighborhood'].apply(list).apply(lambda x:','.join(x)).to_frame().reset_index()
df = df.replace(r'\n',' ', regex=True)
df
```

```python
url="http://cocl.us/Geospatial_data"
s=requests.get(url).content
c=pd.read_csv(io.StringIO(s.decode('utf-8')))
```

```python
c.columns = ['postcode', 'Latitude', 'Longitude']
df = pd.merge(c, df, on='postcode')
df = df[['postcode','borough', 'neighborhood', 'Latitude', 'Longitude']]
df
```

```python
address = 'Toronto, ON'

geolocator = Nominatim(user_agent="yyz_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))

map_tor = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(df['Latitude'], df['Longitude'], df['borough'], df['neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_tor)  
    
map_tor
```
